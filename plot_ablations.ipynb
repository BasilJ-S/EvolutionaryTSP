```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("./Results/2024-12-11_19:07:02.069385/trialResults.csv")

# Group the data by 'Prob of Errors' and 'Population Size'
grouped = df.groupby(['Prob of Errors', 'Population Size']).mean().reset_index()

# Get unique values for population sizes and probabilities
pop_sizes = grouped['Population Size'].unique()
prob_errors = grouped['Prob of Errors'].unique()

x = np.arange(len(pop_sizes))  # the label locations
width = 0.1  # the width of the bars

fig, ax = plt.subplots(layout='constrained')

# Define a color map
colors = plt.cm.plasma(np.linspace(0, 1, len(prob_errors)))

# Plot bars for each probability of error
for i, prob in enumerate(prob_errors):
    data = grouped[grouped['Prob of Errors'] == prob]
    offset = width * i
    rects = ax.bar(x + offset, data['Best Score'], width, label=f'Prob of Mutation {1 - prob:.2f}', color=colors[i])
    ax.bar_label(rects, padding=3, rotation='vertical', fontsize=12)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Best Score', fontsize=14)
ax.set_xlabel('Population Size', fontsize=14)
#ax.set_title('Best Score by Population Size and Probability of Mutation')
ax.set_xticks(x + width * (len(prob_errors) - 1) / 2, pop_sizes)
ax.legend(loc='upper left', ncols=3)
fig.set_size_inches(17, 12)
ax.set_ylim(0, 49000)
plt.show()
```
